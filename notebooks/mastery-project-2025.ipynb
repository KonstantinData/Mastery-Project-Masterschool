Calculated nights ≠ Actual nights spent in the room

It's the number of billable nights based on hotel policy, not necessarily physical presence overnight.

```python
from datetime import datetime, timedelta
import pandas as pd

# Hotel rules
CHECKIN_TIME = datetime.strptime("15:00", "%H:%M").time()
CHECKOUT_TIME = datetime.strptime("11:00", "%H:%M").time()
GRACE_PERIOD = timedelta(hours=1)  # Grace period: 1 hour after official checkout time

# Realistic nights calculation function based on hotel logic
def calculate_nights_realistic(row):
    check_in = row["check_in_time"]
    check_out = row["check_out_time"]

    if pd.isna(check_in) or pd.isna(check_out):
        return None  # Missing values

    check_in_date = check_in.date()
    check_out_date = check_out.date()

    # Base number of nights = difference in calendar days
    nights = (check_out_date - check_in_date).days

    # Checkout deadline including grace period
    checkout_deadline = datetime.combine(check_out_date, CHECKOUT_TIME) + GRACE_PERIOD

    if check_out > checkout_deadline:
        nights += 1  # Late checkout → extra night

    return nights


# 2. Calculate nights using realistic hotel logic
df_hotels["nights_calculated"] = df_hotels.apply(calculate_nights_realistic, axis=1)

# 3. Compare with original 'nights' column
df_hotels["nights_diff"] = df_hotels["nights_calculated"] - df_hotels["nights"]

# 4. Show discrepancies if any
inconsistent = df_hotels[df_hotels["nights_diff"] != 0]

print("✅ First rows with calculated and reported nights:")
print(df_hotels[["check_in_time", "check_out_time", "nights", "nights_calculated", "nights_diff"]].head())

if not inconsistent.empty:
    print("\n⚠️ Bookings with discrepancies:")
    print(inconsistent[["check_in_time", "check_out_time", "nights", "nights_calculated", "nights_diff"]])
else:
    print("\n🎉 All bookings are correctly calculated!")
```

```python
df_cluster = df_flights.groupby('user_id').agg({
    'base_fare_usd': ['mean', 'sum', 'count', 'max', 'min']
})
df_cluster.columns = ['fare_mean', 'fare_sum', 'flights_count', 'fare_max', 'fare_min']
df_cluster = df_cluster.reset_index()
```

```python
from sklearn.preprocessing import StandardScaler

features = ['fare_mean', 'fare_sum', 'flights_count', 'fare_max', 'fare_min']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cluster[features])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=8, random_state=42)
df_cluster['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
df_cluster.groupby('cluster')[features].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=df_cluster, x='cluster', y='fare_sum')
plt.title('Clustervergleich: Gesamtflugkosten je User')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(k_range, inertia, marker='o')
plt.xlabel('Anzahl Cluster (k)')
plt.ylabel('Inertia')
plt.title('Elbow-Methode zur Bestimmung von k')
plt.grid(True)
plt.show()
```

```python
# Erst: Datentypen prüfen
df_flights[['return_flight_booked', 'checked_bags']].dtypes

# Dann: Rückflüge in numerische Werte umwandeln (falls bool oder yes/no)
df_flights['return_flight_booked'] = df_flights['return_flight_booked'].astype(int)

# Gruppieren und mitteln (z. B. wie oft Rückflug gebucht wurde)
df_behavior = df_flights.groupby('user_id').agg({
    'return_flight_booked': 'mean',
    'checked_bags': 'mean'
}).reset_index()

# Mit dem bisherigen Cluster-Dataset mergen
df_cluster = df_cluster.merge(df_behavior, on='user_id', how='left')

# Neue Features in die Skalierung aufnehmen
features_extended = features + ['return_flight_booked', 'checked_bags']
X_scaled_ext = scaler.fit_transform(df_cluster[features_extended])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
components = pca.fit_transform(X_scaled_ext)

df_cluster['pca1'] = components[:, 0]
df_cluster['pca2'] = components[:, 1]

sns.scatterplot(data=df_cluster, x='pca1', y='pca2', hue='cluster', palette='tab10')
plt.title('Cluster-Visualisierung mit PCA')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend(title='Cluster')
plt.show()
```
